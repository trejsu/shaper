```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mimg
from tqdm import tqdm_notebook as tqdm
import os
import numpy as np
import matplotlib.gridspec as gridspec
import imageio
from collections import defaultdict
import math
```

```python
NUM_IMAGES = 10000
NUM_SAMPLES = 20
IMAGES_PATH = '/mnt/users/chrusciel/local/shaper_data/mnist/test_png'
DRAWINGS_PATH = '/mnt/users/chrusciel/local/shaper_data/mnist/test_drawings'
```

```python
images = os.listdir(IMAGES_PATH)
```

```python
images_samples_names = np.random.choice(images, NUM_SAMPLES)
images_samples_ids = [name.split('mnist-')[1].split('-y')[0] for name in images_samples_names]
```

```python
drawings = os.listdir(DRAWINGS_PATH)
len(drawings)
```

```python
drawings_samples_names = [name for name in drawings for i in images_samples_ids if f'-{i}-' in name]
len(drawings_samples_names)
```

```python
drawings_samples_names[0]
```

```python
groupped_by_n = defaultdict(list)
for d in drawings_samples_names:
    action_num = int(d.split('.jpg')[0].split('-')[-1])
    groupped_by_n[action_num].append(d)
```

```python
len(groupped_by_n)
```

```python
def plot_grid(images, path, title):
    arrays = [mimg.imread(os.path.join(path, img)) for img in images]
    grid = np.concatenate(arrays, axis=1)
    plt.figure(figsize=(20, 200))
    plt.axis('off')
    plt.title(title)
    plt.imshow(grid)
```

```python
plot_grid(images_samples_names, IMAGES_PATH, 'mnist')
for n in range(1, 20):
    plot_grid(groupped_by_n[n], DRAWINGS_PATH, )
```
